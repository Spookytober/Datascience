## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task10**:

Describe what processes "generate" each of the previous discrete distributions models. Also report their support.

Here's a spicier way to phrase the question: describe these distributions' maximum entropy conditions in english.

Hint: Make sure to read any of the: preliZ documentation, relevant wikipedia articles, or wolfram alpha.


Binomials: Has a support from 0 to infinity on the x axis. This distribution is generated when you wish to know the probability of something happening within n number of trials.

Poisson: This distribution gives the probability of something happening k times in a certain amount of time. Its support is 0 to infinity in the x axis, because events cannot occur less than 0 times.

Bernoulli: Describes the probability of successes and failures given one single trial. It is generated when one single trial is predicted, and it only is supported at x=0 and x=1.

Geometric: The Geometric distribution is generated when you want to determine the success of doing something for the first time given a certain number of trials. Its support is 0 to infinity.

Hypergeometric: This distribution describes the probability of k successes in n draws without replacing them, in a population size of N. It would have been useful in our expiriment involving red and white marbles if we had not replaced them.

Categorical: It states the probability of a random variable being in distinct categories, and the sum of those probabilities is equal to 1. It has a support of all natural numbers.
