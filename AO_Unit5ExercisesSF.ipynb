## From Linear Model to Generalized Linear Model

You'll now fit a linear regression on the height and weight for people or the Kalahari !Kung San people. Height will be the predictor and weight will be the predicted.

Here's the raw data.

**Task5**:

Create and run a stright line linear regression model on the howell data set with height predicting weight. Visualize your model and its posterior.

```python
with pm.Model() as model_lhowell:
    α = pm.Normal("α", mu=0, sigma=100)
    β = pm.Normal("β", mu=0, sigma=10)
    μ = pm.Deterministic("μ", α + β * howell.height)
    σ = pm.HalfCauchy("σ", 10)
    y_pred = pm.Normal("y_pred", mu=μ, sigma=σ, observed=howell.weight)
    idata_lhowell = pm.sample()
```

```python
posterior = az.extract(idata_lhowell, num_samples=100)
x_plot = xr.DataArray(
    np.linspace(howell.height.min(), howell.height.max(), 50),
    dims="plot_id"
    )
mean_line = posterior["α"].mean() + posterior["β"].mean() * x_plot
lines = posterior["α"] + posterior["β"] * x_plot
_, ax = plt.subplots()
ax.plot(x_plot, lines.T, c="C1", alpha=0.2, label="lines")
ax.plot(x_plot, mean_line, c="C0", label="mean line")
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

**Task6**:

Run a posterior predictive check for model_howell, plot the posterior predictive distribution, and comment on the fit of the model you made in Task5.

```python
pm.sample_posterior_predictive(idata_lhowell, model=model_lhowell,  extend_inferencedata=True)
```

```python
mean_line = idata_lhowell.posterior["μ"].mean(("chain", "draw"))
heights = np.random.normal(howell.height.values, 0.01)
idx = np.argsort(heights)
x = np.linspace(heights.min(), heights.max(), 15)
y_pred_q = idata_lhowell.posterior_predictive["y_pred"].quantile(
    [0.03, 0.97, 0.25, 0.75], dim=["chain", "draw"]
)
y_hat_bounds = iter(
    [
        PchipInterpolator(heights[idx], y_pred_q[i][idx])(x)
        for i in range(4)
    ]
)
_, ax = plt.subplots()
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)
ax.plot(howell.height[idx], mean_line[idx], c="C0")
for lb, ub in zip(y_hat_bounds, y_hat_bounds):
    ax.fill_between(x, lb, ub, color="C1", alpha=0.5)
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

```python
az.plot_ppc(idata_lhowell, num_pp_samples=200, colors=["C1", "C0", "C1"])
```

One major issue that I see with this model is that it goes past zero. By using a normal distribution we have not limited our domain to positive values, meaning that the model is predicting that some short people have negative weights, which we know is not possible. The model did match the high peak on the right fairly well, but a model that is limited to positive numbers could likely model this much better.

**Task7**:

Create and visualize the posterior of a generalized linear model that improves the on the fit of the model you made in Task5.

```python
with pm.Model() as model_nehw:
    α = pm.Normal("α", mu=0, sigma=100)
    β = pm.Normal("β", mu=0, sigma=10)
    μ = pm.Deterministic("μ", pm.math.exp(α + β * howell.height))
    σ = pm.HalfNormal("σ", 10)
    y_pred = pm.Normal("y_pred", mu=μ, sigma=σ, observed=howell.weight)
    idata_nehw = pm.sample()
```

```python
pm.sample_posterior_predictive(idata_nehw, model = model_nehw, extend_inferencedata=True)
```

```python
mean_line = idata_nehw.posterior["μ"].mean(("chain", "draw"))
temperatures = np.random.normal(howell.height.values, 0.01)
idx = np.argsort(temperatures)
x = np.linspace(temperatures.min(), temperatures.max(), 15)
y_pred_q = idata_nehw.posterior_predictive["y_pred"].quantile(
    [0.03, 0.97, 0.25, 0.75], dim=["chain", "draw"]
)
y_hat_bounds = iter(
    [
        PchipInterpolator(temperatures[idx], y_pred_q[i][idx])(x)
        for i in range(4)
    ]
)

_, ax = plt.subplots()
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)
ax.plot(howell.height[idx], mean_line[idx], c="C0")

for lb, ub in zip(y_hat_bounds, y_hat_bounds):
    ax.fill_between(x, lb, ub, color="C1", alpha=0.5)

ax.set_xlabel("Height")
ax.set_ylabel("Weight")
```
